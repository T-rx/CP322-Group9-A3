```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import zero_one_loss
```

```python
train_df = pd.read_csv('mnist_train.csv')
test_df = pd.read_csv('mnist_test.csv')

X_train_full = train_df.iloc[:, 1:].values
y_train_full = train_df.iloc[:, 0].values

X_test = test_df.iloc[:, 1:].values
y_test = test_df.iloc[:, 0].values

# 2. Filter for classes 0 to 4
# You just need to work with samples from class 0 to 4 (5 classes)[cite: 10].
train_mask = y_train_full <= 4
X_train_filtered = X_train_full[train_mask]
y_train_filtered = y_train_full[train_mask]

test_mask = y_test <= 4
X_test_filtered = X_test[test_mask]
y_test_filtered = y_test[test_mask]

# Normalize pixel values (optional but highly recommended for neural networks)
X_train_filtered = X_train_filtered / 255.0
X_test_filtered = X_test_filtered / 255.0

# 3. Create the Validation Set
# You need to consider 10% of training as the validation set[cite: 11].
X_train, X_val, y_train, y_val = train_test_split(
    X_train_filtered, 
    y_train_filtered, 
    test_size=0.10, 
    random_state=42
)

mlp = MLPClassifier(
    hidden_layer_sizes=(16, 32), 
    learning_rate_init=0.001,    
    max_iter=1000,               
    batch_size=100,              
    random_state=42
)

# 5. Train the Model
mlp.fit(X_train, y_train)

# 6. Calculate Errors
# Report training, validation, and testing errors.
train_error = 1 - mlp.score(X_train, y_train)
val_error = 1 - mlp.score(X_val, y_val)
test_error = 1 - mlp.score(X_test_filtered, y_test_filtered)

print(f"Training Error: {train_error:.4f}")
print(f"Validation Error: {val_error:.4f}")
print(f"Testing Error: {test_error:.4f}")
```
